```python
import numpy as np
import pandas as pd
```

```python
diag_df = pd.read_csv('Diag_Sum.csv')
bl_df = pd.read_csv('BL_Change.csv')
```

```python
diag_df.columns
```

```python
diag_df = diag_df[['PTID', 'DXCURREN', 'DIAGNOSIS']]
```

```python
bl_df.columns
```

```python
bl_df = bl_df[['PTID', 'BCPREDX']]
```

```python
bl_df['BCPREDX'].unique()
```

```python
mask = bl_df['BCPREDX'] == -4
bl_df = bl_df[~mask]
```

```python
bl_df = bl_df.dropna()
```

```python
bl_df.head()
```

```python
bl_df['BCPREDX'] = bl_df['BCPREDX'].astype(int)
```

```python
bl_df.head()
```

```python
combined = bl_df.merge(diag_df[['PTID', 'DXCURREN']], left_on='PTID', right_on='PTID')
combined = combined.merge(diag_df[['PTID', 'DIAGNOSIS']], left_on='PTID', right_on='PTID')
```

```python
combined = combined.drop_duplicates()
```

```python
combined.head()
```

```python
conversion = combined.loc[(combined['BCPREDX'] < combined['DXCURREN']) | (combined['BCPREDX'] < combined['DIAGNOSIS'])]
```

```python
conversion = conversion.reset_index()
```

```python
conversion = conversion.drop(['index'], axis=1)
```
